```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Crimes_-_2001_to_present.csv')
```

```python
df.head(1)
```

```python
#Solution1 NOrmal filterting
sol1=df[(df.FBICode == '08B') | (df.FBICode  == '04A') | (df.FBICode  == '04B')]
sol1.head(3)
```

```python
#Solution2 - Shorter way Filter
sol2=df[df.FBICode.isin(['08B', '04A', '04B'])]
sol2.head(3)
```

```python
# Solution 3 Invert the filter:
sol3=df[~df.FBICode.isin(['08B', '04B', '04A'])]
sol3.head(3)
```

```python
#Solution 4 
#With query, you can also do:
FBIcode = ['08B', '04B', '04A']
sol4=df.query('FBICode == @FBIcode')
sol4.head(3)
```

```python
# SOlution 5 invert with:
sol5=df.query('FBICode != @FBIcode')
sol5.head(3)
```
